### Pergunta : 
É esperado que a qualidade do saneamento basico e o nivel de corrupçao impacte o Indice de Desenvolvimento Humano (IDH) de um país, porém nem tudo o que é esperado de fato ocorre. Estaria essas variaveis realmente relacionadas?


# Introduçao:

O Índice de Desenvolvimento Humano (IDH) é uma unidade de medida uttilizada para medir o grau de "desenvolvimento humano" de um país e existem diversas variáveis que afetam esse índice. No meu estudo procurei saber se há relação entre o saneamento basico, a corrupção e o IDH, pois o mais esperado é que em paises corruptos haja menos investimentos na saúde publica e consequentemente um menor valor no IDH.




```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
import matplotlib

```

# Analise exploratoria:

```python
IDH = pd.read_excel('IDH.xlsx')
Saneamento = pd.read_excel('basic_sanitation.xlsx')
Corrupçao = pd.read_excel('Corruption.xlsx')


```

```python
IDH2009 = IDH.loc[:, ['Pais', 2009] ]
Saneamento2009 = Saneamento.loc[:, ['Pais', 2009] ]
Corrupçao2009 = Corrupçao.loc[:, ['Pais', 2009] ]
```

```python
IDH2009 = IDH2009.set_index('Pais')
Saneamento2009 = Saneamento2009.set_index('Pais')
Corrupçao2009 = Corrupçao2009.set_index('Pais')
```

```python
Dataframe =  IDH2009.join(Saneamento2009, how='inner', lsuffix='_IDH', rsuffix='_Saneamento')
Dataframe = Dataframe.join(Corrupçao2009, how='inner')
Dataframe.rename(columns={2009:"2009_Corrupçao"}, inplace=True)






```

### Graficos de disperçao (2009):

```python
IDHxSaneamento = Dataframe.plot.scatter(x="2009_IDH", y = "2009_Saneamento", title = "Saneamento X IDH (2009)")
IDHxCorrupçao = Dataframe.plot.scatter(x="2009_IDH", y = "2009_Corrupçao",title = "Corrupçao X IDH (2009)")
SaneamentoXCorrupçao = Dataframe.plot.scatter(x="2009_Saneamento", y = "2009_Corrupçao", title = "Corrupçao X Saneamento (2009)")
```

```python
Dataframe.corr()
```

# Conclusao:

Para o estudo da pergunta proposta, foram gerados 3 graficos de disperçao e uma tabela de correlaçao, o que facilita indicar a existência, ou não, de relações entre as variáveis escolhidas. 

Primeiramente foi analisado a correlaçao entre o IDH, a porcentagem de pessoas com acesso ao serviços basicos de saneamento e o nível de corrupção dos paises no ano de 2009.

O primeiro grafico exibido foi entre IDH e saneamento basico e conforme observado na figura e no valor de 0,903 de correlaçao, percebe-se uma associaçao forte entre essas variáveis. O segundo grafico foi entre o IDH e a corrupçao em que o valor de correlaçao foi de 0,732 uma relaçao significativa porém não tao forte como a anterior e o ultimo grafico foi entre o saneamento basico e a corrupção, onde o valor de correlaçao foi de 0,553 o menor entre todos.

Ao analisar esses dados, pode-se concluir que realmente há uma forte relaçao entre o saneamento basico e o IDH dos paises, existem estudos que mostram que isso ocorre pois a qualidade do saneamento basico afeta diretamente os critérios utilizados pelo IDH: educação, saúde e renda (http://www.tratabrasil.org.br/blog/2010/11/11/falta-de-saneamento-reflete-no-idh/). Pude concluir também que há uma relaçao significativa entre IDH e o índice de corrupçao de um país, porém nao significa que um país corrupto necessariamente terá um IDH baixo, pois a correlaçao não é tao forte a esse ponto. Ao meu ver isso ocorre pois a corrupção nao afeta diretamente no IDH, pode ser que variáveis ocultas causem alterações tanto em um quanto no outro fazendo assim com que a correlaçao seja consideravelmente alta. A relaçao entre corrupçao e saneamnto também não foi muito alta, contradizendo o que é esperado pelo senso comum, o que nos levam a questionar o motivo pelo qual isso ocorre, o que pode ser investigado em trabalhos futuros.
